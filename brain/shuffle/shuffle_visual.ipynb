```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

plt.figure(figsize=(6, 4))

# Compute x-range for KDE
all_values = np.concatenate([dict_after_shuffle[key_to_plot][i] for i in range(3)])
x_vals = np.linspace(all_values.min(), all_values.max(), 300)

# KDE for NICHEVI (standard line)
nichevi = gaussian_kde(dict_after_shuffle[key_to_plot][0])
plt.plot(
    x_vals,
    nichevi(x_vals),
    label="NICHEVI",
    color="red",
    linestyle=(0, (1, 2)),  # short dashes
    linewidth=2,
)

# KDE for BANKSY (line with square-like dash)
banksy = gaussian_kde(dict_after_shuffle[key_to_plot][2])
plt.plot(
    x_vals,
    banksy(x_vals),
    label="BANKSY",
    color="blue",
    linestyle=(0, (1, 1, 4, 1)),  # custom dash pattern
    linewidth=2,
)

# KDE for SCANVI (circles as marker)
scanvi = gaussian_kde(dict_after_shuffle[key_to_plot][1])
y_vals_scanvi = scanvi(x_vals)
plt.plot(
    x_vals,
    y_vals_scanvi,
    label="SCANVI",
    color="green",
    linestyle="None",
    marker="o",
    markersize=2,
    markevery=3,
)

# Labels, legend, and export
plt.xlabel("cLISI values")
plt.ylabel("Density")
plt.title(f"KDE of {key_to_plot} cLISI - After Shuffle")
plt.legend()
plt.tight_layout()

# plt.savefig(
#     f"{FIGURES_FOLDER}/{dict_after_shuffle[key_to_plot][-1]}_cLISI_cdf.png", dpi=1000
# )
# plt.savefig(
#     f"{FIGURES_FOLDER}/{dict_after_shuffle[key_to_plot][-1]}_cLISI_cdf.svg", dpi=1000
# )

plt.show()
```
